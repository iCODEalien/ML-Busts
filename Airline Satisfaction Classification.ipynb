### There are signs of outliers lets summarize them


```python
# Check for outliers.
def outliers(df,features):
    """Count the number of outliers for each feature using the IQR"""
    num_outliers = []
    pct_outliers = []
    total_rows = []
    interquartile_range = []
    for i in numerical_features:
        Q1 = df[i].quantile(.25)
        Q3 = df[i].quantile(.75)
        IQR = Q3 - Q1
        outliers = len(df[(df[i] < (Q1-1.5*IQR)) | (df[i] > (Q3+1.5*IQR))])
        rows = len(df[i])
        pct = outliers/rows
        interquartile_range.append(IQR)
        num_outliers.append(outliers)
        pct_outliers.append(pct)
        total_rows.append(rows)
        
    count_outliers = pd.DataFrame({'Feature': numerical_features
                               , 'Num_Outliers': num_outliers
                                , 'Percent_Outliers': pct_outliers
                                , 'IQR': interquartile_range
                               , 'Total_Rows': total_rows}).sort_values('Percent_Outliers', ascending=False)
    return count_outliers

count_outliers = outliers(df = df, features = numerical_features)
count_outliers
```

## Treating the Outliers

```python
for i in numerical_features:
    if i != "Numcolumn":
        print("-"*90)
        print(i)
        IQR = df[i].quantile(0.75) - df[i].quantile(0.25)
        UL = df[i].quantile(0.75) + 1.5*IQR
        LL = df[i].quantile(0.25) - 1.5*IQR
        print("IQR:",IQR)
        print("UL: ",UL)
        print("LL:",LL)
        df[i] = np.where(df[i]>UL, UL,df[i])
        df[i] = np.where(df[i]<LL, LL, df[i])
        outliers = len(df[(df[i] < LL) | (df[i] > UL)])
        print("outliers: ",outliers)
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, classification_report, confusion_matrix
```

```python
df.info()
```

```python
df["Class"].value_counts()
```

```python
#__ Scikit learn classifiers cannot handle categorical variables therefore one hot encoding using pd,get_dummies

df["Customer Type"] = df["Customer Type"].replace(["Loyal Customer","disloyal Customer"],[1,0])
df["Gender"] = df["Gender"].replace(["Female","Male"],[0,1])
df["Type of Travel"] = df["Type of Travel"].replace(["Business travel","Personal Travel"],[1,0])

one_hot_encoded_variables = pd.get_dummies(df["Class"],prefix='Class', drop_first = "True")
print(one_hot_encoded_variables)
df = pd.concat([df,one_hot_encoded_variables], axis = 1)
df.head()
```

```python
df["Type of Travel"].value_counts()
```

```python
#_______Basic tree model

X = df.drop(["satisfaction","Class"], axis = 1)
y = df["satisfaction"].copy()
```

```python
y = y.replace(["neutral or dissatisfied","satisfied"],[0,1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
y_pred
```

```python
#___-Confusion matrix cross tab
pd.crosstab(y_test,y_pred, margins = True,rownames = ['Actual'], colnames = ['Predicted'] ,margins_name = "All")
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#_accuracy score
accuracy_score(y_test,y_pred)
```

```python
from sklearn import tree

tree.plot_tree(clf.fit(X_train,y_train))
```

```python
# Pruning
ccp_alphas = []
prun = clf.cost_complexity_pruning_path(X_train,y_train)
ccp_alphas = prun['ccp_alphas']
ccp_alphas[0:20]
```

```python
accuracy_i = []
```

```python
for i in ccp_alphas:
    clf = DecisionTreeClassifier(ccp_alpha = i)
    clf.fit(X_train,y_train)
    y_pred_test = clf.predict(X_test)
    accuracy_i.append(accuracy_score(y_test,y_pred_test))
```

```python
sns.set()
plt.figure(figsize=(16,10))
sns.lineplot(y=accuracy_i,x=ccp_alphas,label='Accuracy')
plt.xticks(ticks=np.arange(0.00,0.05,0.01))
plt.xlabel('alphas')
plt.ylabel('accuracy')
plt.title('CCP_Alpha v/s accuracy')
plt.show()
```

```python
## Finding the least accuracy for ccp_alpha
ccp_series = pd.Series(ccp_alphas)
accuracy_series = pd.Series(accuracy_i)
A_R = pd.concat([ccp_series,accuracy_series], axis = 1)
A_R = A_R.rename(columns = {0:'ccp_alpha',1:'accuracy'})
A_R.head()
```

```python
print("Lowest accuracy value for ccp_alphas: \n",A_R[A_R['accuracy'] == A_R['accuracy'].max()])
```
