## Outlier identification

```python
from scipy import stats
for col in df.columns:
    iqr=stats.iqr(df[col])
    print("\nIQR of ",col,":",iqr)
    
    upper_limit = df[col].quantile(0.75)+1.5*iqr
    print("No of outliers in",col,":", len(df[df[col]>upper_limit]))
```

```python
#_____OUTLIER TREATMENT using Z score method

for i in df.columns:
    z = []
    z_score = stats.zscore(df[i])
    print(z_score)
    z.append(z_score)
```

```python
z_scores = stats.zscore(df)

abs_z_scores = np.abs(z_scores)

new_values = (abs_z_scores < 3).all(axis=1)

new_df = df[new_values]

print(new_df)
```

```python
df1 = pd.DataFrame(tran_val)
df1.head()
```

```python
df1.set_index(new_df.index,inplace = True)
df1.head()
```

# Finding the optimum number of clusters

```python
from sklearn.cluster import KMeans
```

```python
# Within cluster Sum of Squares(WCSS)
wcss =[]

for i in range(1,12):
    kmeans = KMeans(n_clusters = i, random_state = 42, init = 'random')
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,12),wcss,'-gx',)
plt.title("Optimum no of clusters")
plt.xlabel('Number of clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
#____seems like optimum number of clusters is 3
kmeans = KMeans(n_clusters = 3)
cluster = kmeans.fit_predict(df1)
print(cluster)
```

```python
fc = cluster+1
cluster = list(fc)
new_df['cluster'] = cluster
new_df.head()
```

# Silhouette scores and samples

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
```

```python
# Overall Silhouette Score
print(f'silhouette score(clusters = 4): {silhouette_score(df1,cluster)}')
```

## plotting Silhouette Score

```python
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
for i in range(2,11):
    model = KMeans(n_clusters = i, random_state = 12)
    visualizer = SilhouetteVisualizer(model, colors = 'yellowbrick')
    visualizer.fit(df1)
    visualizer.show()
```

```python
cluster_labels = np.unique(clusters)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(df1,clusters,metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []

for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[clusters == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper),
             c_silhouette_vals,
             height=1.0,
             edgecolor='none',
             color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg,
            color="red",
            linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

#data.groupby('cluster').mean()
```
